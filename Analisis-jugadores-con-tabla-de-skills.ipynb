```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
Valioso=pd.read_excel('Pepechum.xlsx',sheet_name='pepe')
BIGDATA=pd.read_excel('Pepechum.xlsx',sheet_name='BIGDATA')
```

```python
Valioso.groupby('Nombre').sum().sort_values(by='global',ascending=False)['global'].head(50)
Valioso.groupby('Nombre').sum().sort_values(by='ofensivo',ascending=False)['ofensivo'].head(50)
Valioso.groupby('Nombre').sum().sort_values(by='defensivo',ascending=False)['defensivo'].head(50)
Valioso.keys()
```

```python
#ANALISIS DE DEFENSORES 
lista=['Entradas','Pases acertados', 'Porc Efectividad pases','Duelos ganados','Duelos en el suelo ganados',
 'Duelos aereos ganados',  'Despejes','Disparos Bloq Def',
 'Intercepciones', 'Toques','Pases clave', 'Balones largos completados']
listanegativa=['Faltas', 'Regateado','Posesion perdida']
Listado=[]
ListadoN=[]
Atributos=[]
atr=[1]
for a in lista:
    l=list(BIGDATA.loc[BIGDATA['Posicion']=='D'].groupby('Nombre').sum().sort_values(by=a,ascending=False).head(5).index)
    Listado=Listado+l
for a in listanegativa:
    l=list(BIGDATA.loc[BIGDATA['Posicion']=='D'].groupby('Nombre').sum().sort_values(by=a,ascending=False).head(5).index)
    ListadoN=ListadoN+l
#print(Listado)
df=pd.DataFrame(columns=['Jugador','Cantidad'])
dfN=pd.DataFrame(columns=['JugadorN','CantidadN'])
df['Jugador']=Listado
df['Cantidad']=1
dfN['JugadorN']=ListadoN
dfN['CantidadN']=-1
print(df.groupby('Jugador').sum().sort_values(by='Cantidad',ascending=False).head(20))
print(dfN.groupby('JugadorN').sum().sort_values(by='CantidadN',ascending=True).head(20))
BIGDATA.loc[BIGDATA['Nombre']==df.groupby('Jugador').sum().sort_values(by='Cantidad',ascending=False).index[0]]
for b in range(0,5) :
    for a in lista:
        l=list(BIGDATA.loc[BIGDATA['Posicion']=='D'].groupby('Nombre').sum().sort_values(by=a,ascending=False).head(5).index)
        if df.groupby('Jugador').sum().sort_values(by='Cantidad',ascending=False).index[b] in l:
            atr[0]=a
            Atributos=Atributos+atr
    print(df.groupby('Jugador').sum().sort_values(by='Cantidad',ascending=False).index[b]+' '+ str(Atributos))
    print(' ')
    Atributos=[]
```
